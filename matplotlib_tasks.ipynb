```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Name": ["Madina", "Isak", "Zarina", "Neila", "Nurislam"],
    "Score": [85, 90, 78, 92, 88]
}
df = pd.DataFrame(data)
```

```python
plt.plot(df['Name'], df['Score'], marker='o')
plt.xlabel("Student Name")
plt.ylabel("Score")
plt.title("Student Scores Line Plot")
plt.grid(True)
plt.show()
```

```python
plt.bar(df['Name'], df['Score'], color='skyblue')
plt.xlabel("Student Name")
plt.ylabel("Score")
plt.title("Student Scores Bar Chart")
plt.show()
```
