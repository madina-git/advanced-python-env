```python
import numpy as np
array_1d = np.arange(1, 11)
print("1D Array:", array_1d)
```

```python
array_2d = np.random.randint(1, 11, size=(3,3))
print("2D Array:\n", array_2d)
```

```python
sum_val = array_1d.sum()
mean_val = array_1d.mean()
max_val = array_1d.max()

print("Sum:", sum_val)
print("Mean:", mean_val)
print("Max:", max_val)

array_multiplied = array_1d * 2
print("Array multiplied by 2:", array_multiplied)

array_replaced = np.where(array_1d > 5, 0, array_1d)
print("Array with values >5 replaced by 0:", array_replaced)
```
